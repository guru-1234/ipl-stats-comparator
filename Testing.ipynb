```python
current=[]
curr=[]
for i in hyperlink_2021:
    http = urllib3.PoolManager()
    url = i
    response = http.request('get',url)
    soup_2021 = BeautifulSoup(response.data)
    team_2021 = soup_2021.find('td', {'colspan':'2'}).text
    if "*" in team_2021:
            sed= team_2021.split("*")
            if "\r\n\t\t\t\t\t\t\t\t\t" in sed[0]:
                pain= sed[0].split("\r\n\t\t\t\t\t\t\t\t\t")
                current.append(pain[1])
```

```python
Strike_rate=[]
for i in hyperlink_2021:
    http_ = urllib3.PoolManager()
    url_= i
    response_= http.request('get',url_)
    soup_2021_ = BeautifulSoup(response_.data)
    team_2021_= soup_2021_.find('table', {'cellpadding':'1','width':'270','border':'0'}).text
    cut= team_2021_.splitlines()
    size = len(cut)
    idx_list = [idx + 1 for idx, val in
                 enumerate(cut) if val == "Scoring Rate"]
    res = [cut[i: j] for i, j in
        zip([0] + idx_list, idx_list + 
        ([size] if idx_list[-1] != size else []))]
    
    cut1=[]
    for i in res[1]:
          if "\t\t\t\t\t\t\t\t\t\t\t\t" in i:
                sed_=i.split("\t\t\t\t\t\t\t\t\t\t\t\t")
                cut1.append(sed_[1])
    Strike_rate.append(cut1[0])
```

```python
for i,n in enumerate(Strike_rate):
    if n== "N/A":
        Strike_rate[i]=0
```
